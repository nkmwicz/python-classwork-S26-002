```python
import pandas as pd
import scipy.stats as stats
# https://raw.githubusercontent.com/dataprofessor/
# data/refs/heads/master/penguins_cleaned.csv
url = "https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/penguins_cleaned.csv"
df = pd.read_csv(url)
df.head(2)
```

```python
df.corr(numeric_only=True)
```

```python
r, p = stats.pearsonr(df["body_mass_g"], df["flipper_length_mm"])
r2 = r**2
print(f"r={r:.4f}, p={p:.4f}, r2={r2:.4f}")
```

```python
r, p = stats.pearsonr(df["flipper_length_mm"], df["bill_length_mm"])
r2 = r**2
print(r,p,r2)
```
