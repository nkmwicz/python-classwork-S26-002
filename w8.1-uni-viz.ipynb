```python
# %conda install seaborn
# %conda install matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/nkmwicz/data-for-students/refs/heads/main/soccer-players.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Annatomy of a Viz
plt.figure(figsize=(5,3))
plt.xlabel("Heights in CM")
plt.ylabel("Frequency")
ymax = 100
xmax = 25
plt.ylim(0,ymax)
plt.xlim(0,xmax)
plt.title("Histogram of Soccer Player Heights (cm)")
string = "center"
# plt.text(x=12.5, y=50, s=string, ha="left", va="bottom")
plt.text(
    x=0.15, 
    y=.90, 
    s=string, 
    ha="left", 
    va="bottom",
    # bbox=dict(boxstyle="round, pad=0.5", fc="white"),
    bbox={"boxstyle": "round, pad=0.5", "fc":"white"},
    transform=plt.gca().transAxes)
plt.axvline(x=xmax/2, color="r", linestyle="--", label="Vertical")
plt.axhline(y=ymax/2, color="blue", linestyle="--", label="Horizontal")
plt.grid()
plt.legend()
plt.show()
```

```python
import scipy.stats as stats
```

```python
# Histogram with plt
data = df["height_in_cm"].dropna()

plt.figure(figsize=(5,3))
bins = round(data.count() ** (1/3))
plt.hist(x=data, bins=bins, edgecolor="black")
plt.xlabel("Heights in CM")
plt.ylabel("Frequency")
ymax = 100
xmax = 25
plt.title("Histogram of Soccer Player Heights (cm)")
sk = stats.skew(data)
kurt = stats.kurtosis(data)
string = f"""Skew: {sk:.2f}
Kurtosis: {kurt:.2f}"""
# plt.text(x=12.5, y=50, s=string, ha="left", va="bottom")
plt.text(
    x=0.20, 
    y=.80, 
    s=string, 
    ha="center", 
    va="center",
    # bbox=dict(boxstyle="round, pad=0.5", fc="white"),
    bbox={"boxstyle": "round, pad=0.5", "fc":"white"},
    transform=plt.gca().transAxes)
plt.axvline(x=data.mean(), color="r", linestyle="--", label="Mean")
plt.axvline(x=data.median(), color="b", linestyle="--", label="Median")
plt.grid()
plt.legend()
plt.show()
```
