```python
from scipy import stats
pd_skew = df["height_in_cm"].skew()
scipy_skew = stats.skew(df["height_in_cm"], nan_policy="omit")
print(f"pd_skew={pd_skew}; scipy_skew ={scipy_skew}")
```

```python
pd_kurt = df["height_in_cm"].kurt()
scipy_kurt = stats.kurtosis(df["height_in_cm"].dropna())
scipy_kurt_omit = stats.kurtosis(df["height_in_cm"], nan_policy="omit")

print(f"pd_kurt = {pd_kurt}\nscipy_kurt={scipy_kurt}\nscipy_kurt_omit={scipy_kurt_omit}")
```

```python
import pandas as pd

def eda(df: pd.DataFrame) -> None:
    """
    Do stats eval on dataframe
    
    Args:
        df (pd.DataFrame): Dataframe to evaluate
    Returns:
        None
    """
    num_cols = []
    cat_cols = []
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            num_cols.append(col)
        else:
            cat_cols.append(col)
    print(f"{'='*5}DF SHAPE: {df.shape}{'='*5}")
    print(f"{'*'*3}Numeric Columns{'*'*3}")
    for col in num_cols:
        print(f"{col}")
        max = df[col].max()
        min = df[col].min()
        print(f"\tMin = {min}")
        print(f"\tMax = {max}")
        print(f"\tRange = {max - min}")
```

```python
eda(df)
```
