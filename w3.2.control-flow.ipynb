```python
# Conditionals: ==, <, <=, >, >=, !=

a:int = 10
b:int = 12

print(f"a={a}, b={b}")
print(f"a==b: {a==b}")
print(f"a>b: {a>b}")
print(f"a<=b: {a<=b}")
```

```python
from typing import List, Dict, Tuple, Set
# type hinting: str, int, float, bool
# in, or, and, is

c: List[int] = [1,3,10,6,7] # for multiples List[str | int]

# in operator
print(f"is {a} in {c}: {a in c}")

# and operator
print(a==b and a in c) # cannot do this: a > b and 5

# or operator
print(a==b or a in c)

# is operator
print(type(a) is int)

# print(type(a) == int)

print(a != b)

# not
print(not True)
print(f"not false: {not False}")
print(f"a<b: {a<b}")
print(f"not a < b: {not a<b}")
```

```python
# truthy / falsy

# falsy = 0, 0.0, "", None, [], {}, ()

print(bool(a))
print(bool(b))
print(bool(c))
print(bool(0))
print(bool([]))
```

```python
# try except
try:
    num: str = input("Enter a number b/w 1 & 10")
    if num.isnumeric():
        num = int(num)
        if num < 1 or num > 10:
            print("yey")
        else:
            raise Exception("Follow Directions. I said between 1 and 10")
    else:
        raise Exception("Choose a number!")
except Exception as e:
    print(f"There was an error: {e}")
```

```python
num: str = input("Enter a number b/w 1 & 10 or 20 & 30")

try:
    if not num.isnumeric():
        raise Exception("Choose a number!")
    num = int(num)
    if (num <1 or num > 10) and (num <20 or num > 30):
        raise Exception("Number should existe between 1 and 10 or 20 and 30")
    
    print(f"{num}  is following the rules")
except Exception as e:
    print(e)
```
